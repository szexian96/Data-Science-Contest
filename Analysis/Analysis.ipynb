## 2. Daily Product

## 2.1 Clean Daily Product

### 2.1.1 Clean unwanted column

```python
df_daily = df_daily.drop(columns=['商品コード','カテゴリ1コード','カテゴリ2コード'])
df_daily.head()
```

### 2.1.2 Clean Date Format & Change to date format

```python
def date_remove(date):
    return date.replace('/', '')

def day_remove(day):
    return day.split('(')[0]

df_daily['Date'] = df_daily['Date'].apply(lambda x: day_remove(x))
df_daily['Date'] = df_daily['Date'].apply(lambda x: date_remove(x))
df_daily['Date'] = pd.to_datetime(df_daily['Date']) #change the format to datetime
df_daily
```

### 2.1.3 Clean Price and Total then change to numeric

```python
def yen_remove(yen):
    return yen.replace('¥', '')

def remove(comal):
    return comal.replace(',', '')    

df_daily['Price'] = df_daily['Price'].apply(lambda x: yen_remove(x))
df_daily['Price'] = df_daily['Price'].apply(lambda x: remove(x))

df_daily['Total'] = df_daily['Total'].apply(lambda x: yen_remove(x))
df_daily['Total'] = df_daily['Total'].apply(lambda x: remove(x))

df_daily['Price'] = pd.to_numeric(df_daily['Price'])
df_daily['Total'] = pd.to_numeric(df_daily['Total'])
```

### 2.1.4 Change name for product and type

```python
def l_remove(l):
    return l.replace(')','')

def l2_remove(l):
    return l.replace('(','')

def remove(l):
    return l.replace('L','')

def to_remove(l):
    return l.replace('TO','')

def set_remove(l):
    return l.replace('ｾｯﾄ','')

def d_remove(l):
    return l.replace('D','')

def af_remove(l):
    return l.replace('AF','')

def space_remove(l):
    return l.replace(' ','')

df_daily['Product'] = df_daily['Product'].apply(lambda x: l_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: l2_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: to_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: set_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: d_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: af_remove(x))
df_daily['Product'] = df_daily['Product'].apply(lambda x: space_remove(x))

df_daily['Type'] = df_daily['Type'].apply(lambda x: l_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: l2_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: to_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: set_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: d_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: af_remove(x))
df_daily['Type'] = df_daily['Type'].apply(lambda x: space_remove(x))

df_daily
#df_daily.to_csv('clean_daily_all.csv',index=False)
```
